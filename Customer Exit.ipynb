```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head(10)
```

```python
dataset_dummies = pd.get_dummies(data=dataset, columns=['Geography', 'Gender'],drop_first=True)
dataset_dummies.head(10)
```

```python
y = dataset_dummies.iloc[:, 11].values
dataset_dummies.drop(["Exited"], axis=1, inplace=True)
dataset_dummies.head(10)
```

```python
X = dataset_dummies.iloc[:, 3:].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 8, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
y_pred_category = (y_pred > 0.5)
cm = confusion_matrix(y_test, y_pred_category)
cm
```
